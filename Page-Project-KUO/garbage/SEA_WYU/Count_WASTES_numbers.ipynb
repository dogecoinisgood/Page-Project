```python
import xml.etree.ElementTree as ET
import pandas as pd

# 從檔案載入並解析 XML 資料
tree = ET.parse("/content/drive/MyDrive/Colab Notebooks/dataset_10000_update_0912.xml")
root = tree.getroot()
```

```python
for i in range(10):
    # print(child.tag, child.attrib.get("name"), child[0].attrib.get("label"))
    print(root[i].attrib["name"], root[i][0].attrib["label"])
```

```python
for child in root:
    # print(child.tag, child.attrib.get("name"), child[0].attrib.get("label"))
    print(child.tag, child.attrib["name"], child[0].attrib["label"])
```

```python
# 第一種寫法
df= pd.DataFrame([[image.attrib["name"], image[0].attrib["label"]] for image in root], columns=["name","label"])
# 第二種寫法
df= pd.DataFrame([[image.attrib["name"], image.find(".//polygon").attrib["label"]] for image in root], columns=["name","label"])
df.head
```
